```python
sizes = [10,20,30,40,50,100,200,300,400,500,600,700,800,900,1000]
```

```python
mgr = np.load('./NBoutputs/MG_MMranks.npy')

fig = plt.figure(figsize=(7,5))
ax = plt.gca()
ax.plot([x for x in range(50,1050,50)],np.flip(mgr),'.g-',label='$MG$')
ax.set_xlabel('Model Size $N$')
ax.set_ylabel('$\mathrm{rank}(C_N)$')
plt.legend()
plt.show()
```

```python
single_mgr = np.squeeze(np.load('./NBoutputs/singleseed_mmMGranks.npy'))
multi_mgr = np.squeeze(np.load('./NBoutputs/multiseed_mmMGranks.npy'))
```

```python
fig = plt.figure(figsize=(14,5))
#ax = plt.gca()
ax1 = plt.subplot(121)
ax1.plot(sizes[:5],single_mgr[:5],'.m-',label='Single-seed $MG$')
ax1.plot(sizes[:5],multi_mgr[:5],'.g-',label='Multi-seed $MG$')
ax1.set_title('Small models')
ax1.set_xlabel('Model Size $N$')
ax1.set_ylabel('$\mathrm{rank}(C_N)$')
ax1.legend()



ax2 = plt.subplot(122)
ax2.plot(sizes[5:],single_mgr[5:],'.m-',label='Single-seed $MG$')
ax2.plot(sizes[5:],multi_mgr[5:],'.g-',label='Multi-seed $MG$')
ax2.set_title('Large models')
ax2.set_xlabel('Model Size $N$')
ax2.set_ylabel('$\mathrm{rank}(C_N)$')
ax2.legend()
plt.show()
fig.savefig("./NBoutputs/new_multimodel_ranks.pdf", format='pdf', bbox_inches='tight', pad_inches=.01)
```

```python
single_mger = np.load('./NBoutputs/singleseed_mmMG.npy')
multi_mger = np.load('./NBoutputs/multiseed_mmMG.npy')

fig = plt.figure(figsize=(14,5))
ax1 = plt.subplot(121)
ax1.set_ylim(0,.2)
ax1.plot(sizes[:5],single_mger[:5],'.m-',label='Single-seed $MG$')
ax1.plot(sizes[:5],multi_mger[:5],'.g-',label='Multi-seed $MG$')
ax1.set_title('$MG$ Small Model Error')
ax1.set_xlabel('Model Size $N$')
ax1.set_ylabel('NRMSE $\mathcal{E}$')
ax1.legend()

ax2 = plt.subplot(122)
ax2.set_ylim(0,.2)
ax2.plot(sizes[5:],single_mger[5:],'.m-',label='Single-seed $MG$')
ax2.plot(sizes[5:],multi_mger[5:],'.g-',label='Multi-seed $MG$')
ax2.set_yticks([])
ax2.yaxis.set_visible(False)
ax2.set_ylabel([])
ax2.set_title('$MG$ Large Model Error')
ax2.set_xlabel('Model Size $N$')
#ax2.set_ylabel('NRMSE $\mathcal{E}$')
ax2.legend()
plt.subplots_adjust(wspace=.01)
plt.show()
fig.savefig("./NBoutputs/new_multimodel_error.pdf", format='pdf', bbox_inches='tight', pad_inches=.01)
```

```python
mgr = np.load('./NBoutputs/MG_MMranks.npy')
mgerrors = np.load('./NBoutputs/MG_mm_errors.npy')
```

```python
gs = gridspec.GridSpec(2, 1, hspace=0)
fig = plt.figure(figsize=(10,14))

ax1 = plt.subplot(gs[0])
ax1.plot([x for x in range(50,1050,50)],np.flip(mgr),'.g-',label='$MG$')
ax1.set_xlabel([])
ax1.set_ylabel('$\mathrm{rank}(C_N)$')
ax1.legend(loc='upper left',
           framealpha=1,
           handlelength=.5,
           ncol=2,
           fontsize=20,
           borderpad=.2,
           borderaxespad=.2)

ax2 = plt.subplot(gs[1])
ax2.plot([x for x in range(50,1050,50)],np.flip(mgerrors),'.b-',label='$MG$')

ax2.set_xlabel('Model Size $N$')
ax2.set_ylabel('NRMSE $\mathcal{E}$')

ax2.legend(loc='upper right',
           framealpha=1,
           handlelength=.5,
           ncol=2,
           fontsize=20,
           borderpad=.2,
           borderaxespad=.2)
plt.show()
fig.savefig("./NBoutputs/multimodel_plot.pdf", format='pdf', bbox_inches='tight', pad_inches=.01)
```
