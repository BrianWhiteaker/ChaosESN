```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import time

from sklearn.metrics import mean_squared_error
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.gridspec as gridspec

fontsize = 24
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["text.usetex"] = True 
plt.rcParams["axes.xmargin"] = 0 
plt.rcParams["axes.titlesize"] = 26 
plt.rcParams["axes.labelsize"] = 26 
plt.rcParams["xtick.labelsize"] = fontsize 
plt.rcParams["ytick.labelsize"] = fontsize
plt.rcParams["xtick.color"] = 'black'
plt.rcParams["ytick.color"] = 'black'
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["text.latex.preamble"] = r"\usepackage{amsmath}"  #lmodern
```

```python
l3dr = np.load('.NBout/L3D_MMranks.npy')
lxr = np.load('./Lx_MMranks.npy')
lzr = np.load('./Lz_MMranks.npy')
mgr = np.load('./MG_MMranks.npy')

fig = plt.figure(figsize=(7,5))
ax = plt.gca()

ax.plot([x for x in range(50,1050,50)],np.flip(l3dr),
         '>b',linestyle='dotted',label='$L3D$')
ax.plot([x for x in range(50,1050,50)],np.flip(lxr),'.y-',label='$Lx$')
ax.plot([x for x in range(50,1050,50)],np.flip(lzr),'.m-',label='$Lz$')
ax.plot([x for x in range(50,1050,50)],np.flip(mgr),'.g-',label='$MG$')
ax.set_xlabel('Model Size $N$')
ax.set_ylabel('$\mathrm{rank}(C_N)$')
plt.legend()
plt.show()
```

```python
l3dr = np.squeeze(np.load('./NBoutputs/mmL3Dranks.npy'))
lxr = np.squeeze(np.load('./NBoutputs/mmXranks.npy'))
lzr = np.squeeze(np.load('./NBoutputs/mmZranks.npy'))
mgr = np.squeeze(np.load('./NBoutputs/mmMGranks.npy'))
sizes = [10,20,30,40,50,100,200,300,400,500,600,700,800,900,1000]

fig = plt.figure(figsize=(14,5))
#ax = plt.gca()
ax1 = plt.subplot(121)
ax1.plot(sizes[:5],l3dr[:5],'.b-',label='$L3D$')
ax1.plot(sizes[:5],lxr[:5],'.y-',label='$Lx$')
ax1.plot(sizes[:5],lzr[:5],'.m-',label='$Lz$')
ax1.plot(sizes[:5],mgr[:5],'.g-',label='$MG$')
ax1.set_title('Small models')
ax1.set_xlabel('Model Size $N$')
ax1.set_ylabel('$\mathrm{rank}(C_N)$')
ax1.legend()



ax2 = plt.subplot(122)
ax2.plot(sizes[5:],l3dr[5:],'.b-',label='$L3D$')
ax2.plot(sizes[5:],lxr[5:],'.y-',label='$Lx$')
ax2.plot(sizes[5:],lzr[5:],'.m-',label='$Lz$')
ax2.plot(sizes[5:],mgr[5:],'.g-',label='$MG$')
ax2.set_title('Large models')
ax2.set_xlabel('Model Size $N$')
ax2.set_ylabel('$\mathrm{rank}(C_N)$')
ax2.legend()
plt.show()
fig.savefig("./NBoutputs/new_multimodel_ranks.pdf", format='pdf', bbox_inches='tight', pad_inches=.01)
```

```python
gs = gridspec.GridSpec(2, 1, hspace=0)
fig = plt.figure(figsize=(10,14))

ax1 = plt.subplot(gs[0])
ax1.plot([x for x in range(50,1050,50)],np.flip(l3dr),
         '>b',linestyle='dotted',label='$L3D$')
ax1.plot([x for x in range(50,1050,50)],np.flip(lxr),'.y-',label='$Lx$')
ax1.plot([x for x in range(50,1050,50)],np.flip(lzr),'.m-',label='$Lz$')
ax1.plot([x for x in range(50,1050,50)],np.flip(mgr),'.g-',label='$MG$')
ax1.set_xlabel([])
ax1.set_ylabel('$\mathrm{rank}(C_N)$')
ax1.legend(loc='upper left',
           framealpha=1,
           handlelength=.5,
           ncol=2,
           fontsize=20,
           borderpad=.2,
           borderaxespad=.2)

ax2 = plt.subplot(gs[1])
linestyle = ['+b','<b','>b']
l3label = [f'$L3D \, v^{(1)}$',
           f'$L3D \, v^{(2)}$',
           f'$L3D \, v^{(3)}$']
for i in range(3):
    ax2.plot([x for x in range(50,1050,50)],np.flip(l3der[:,i]),
             linestyle[i],linestyle='dotted', alpha=.5,label=l3label[i])
ax2.plot([x for x in range(50,1050,50)],np.flip(lxer),'.y-',label='$Lx$')
ax2.plot([x for x in range(50,1050,50)],np.flip(lzer),'.m-',label='$Lz$')
ax2.plot([x for x in range(50,1050,50)],np.flip(mger),'.g-',label='$MG$')

ax2.set_xlabel('Model Size $N$')
ax2.set_ylabel('NRMSE $\mathcal{E}$')

ax2.legend(loc='upper right',
           framealpha=1,
           handlelength=.5,
           ncol=2,
           fontsize=20,
           borderpad=.2,
           borderaxespad=.2)
plt.show()
fig.savefig("../../ChaosPlots/multimodel_plot.pdf", format='pdf', bbox_inches='tight', pad_inches=.01)
```
