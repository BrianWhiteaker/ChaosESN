```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import time

from sklearn.metrics import mean_squared_error
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.gridspec as gridspec
import jupyterthemes as jt

fontsize = 24
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["text.usetex"] = True 
plt.rcParams["axes.xmargin"] = 0 
plt.rcParams["axes.titlesize"] = 26 
plt.rcParams["axes.labelsize"] = 26 
plt.rcParams["xtick.labelsize"] = fontsize 
plt.rcParams["ytick.labelsize"] = fontsize
plt.rcParams["xtick.color"] = 'black'
plt.rcParams["ytick.color"] = 'black'
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["text.latex.preamble"] = r"\usepackage{amsmath}"  #lmodern
```

```python
sizes = [10,20,30,40,50,100,200,300,400,500,600,700,800,900,1000]

mgR = np.squeeze(np.load('./NBoutputs/RmmMG.npy'))
mgER = np.load('./NBoutputs/ERRmmMG.npy')
mgCD = np.load('./NBoutputs/CDmmMG.npy')
```

```python
fig = plt.figure(figsize=(14,14))
ax1 = plt.subplot(221)
ax1.set_ylim(0,60)
ax1.plot(sizes[:5],
         mgR[:5],
         '.g-',label='Rank',
         markersize=10)
ax1.set_xlabel('Model Size $N$')
ax1.set_ylabel('$\mathrm{rank}(\mathcal{C}_N)$')
ax1.legend(fontsize=22)
ax1.text(0.05, 0.9,
         '(a1) Small $N$ Rank', fontweight=1000, 
         transform=ax1.transAxes,
         ha="left", fontsize=24)

ax2 = plt.subplot(222)
ax2.set_ylim(0,60)
ax2.plot(sizes[5:],
         mgR[5:],
         '.g-',label='Rank',
         markersize=10)
ax2.text(0.05, 0.9,
         '(a2) Large $N$ Rank', 
         transform=ax2.transAxes, fontweight=1000,
         ha="left", fontsize=24)
ax2.set_yticks([])
ax2.yaxis.set_visible(False)
ax2.set_ylabel([])
ax2.set_xlabel('Model Size $N$')

ax2.legend(fontsize=22)
plt.subplots_adjust(wspace=.07, hspace=.2)

ax3 = plt.subplot(223)
ax3.set_ylim(0,.2)
ax3.plot(sizes[:5],
         mgER[:5],
         '.m-',label='$\mathcal{E}$',
         markersize=10)
ax3.plot(sizes[:5],
         mgCD[:5],
         '.b-',label='CosDist',
         markersize=10)
ax3.text(0.05, 0.9,
         '(b1) Small $N$ Error', 
         transform=ax3.transAxes, fontweight=1000,
         ha="left", fontsize=24)
ax3.set_xlabel('Model Size $N$')
ax3.set_ylabel('NRMSE $\mathcal{E}$')
ax3.legend(fontsize=22, loc='upper right')

ax4 = plt.subplot(224)
ax4.set_ylim(0,.2)
ax4.plot(sizes[5:],
         mgER[5:],
         '.m-',label='$\mathcal{E}$',
         markersize=10)
ax4.plot(sizes[5:],
         mgCD[5:],
         '.b-',label='CosDist',
         markersize=10)
ax4.set_yticks([])
ax4.yaxis.set_visible(False)
ax4.set_ylabel([])
ax4.text(0.05, 0.9,
         '(b2) Large $N$ Error', 
         transform=ax4.transAxes, fontweight=1000,
         ha="left", fontsize=24)
ax4.set_xlabel('Model Size $N$')
ax4.legend(fontsize=22)
plt.show()
fig.savefig("./NBoutputs/mm_rank_and_error.pdf", format='pdf', bbox_inches='tight', pad_inches=.01)
```
